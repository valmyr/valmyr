```python
import matplotlib.pyplot as plt
from math import factorial, pi
def mycos(n):
  s = 0
  for i in range(30):
      s += (pow(-1,i)*pow(n,2*i))/factorial(2*i)
  return s
```

```python
X = []
Y = []
j = -10
while(j < 10):
 Y.append(2*mycos(j))
 X.append(j)
 j+=.1
plt.plot(X,Y)
```
