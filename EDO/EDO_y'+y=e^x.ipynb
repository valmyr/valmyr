```python
import matplotlib.pyplot as plt
from math import factorial, exp
```

```python
def y(x,c1):
  s1 = 0
  s2 = 0
  for i in range(40):
    s1+=(c1*pow(x,2*i))/factorial(2*i)
    s2+=((1-c1)*pow(x,2*i+1))/factorial(2*i+1)
  return s1+s2

  
```

```python
X = []
Y = []
Z = []

x = -1
while(x < 2):
  X.append(x)
  Y.append(y(x,2))
  Z.append(2*(exp(-x)+exp(x/2)))
  x+=0.1
plt.plot(X,Y,color = 'red')
plt.xlabel("Função aproximada Eixo X")
plt.ylabel("Função aproximada Eixo Y")
plt.grid()
plt.show()
plt.plot(X,Z,color = 'blue')
plt.xlabel("Função real Eixo X")
plt.ylabel("Função real Eixo Y")
plt.grid()
plt.show()
```
