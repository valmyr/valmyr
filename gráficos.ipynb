```python
from matplotlib import pyplot as plt
import numpy as np
from math import cos,sin,exp
```

```python
f = lambda h: (np.exp(-x*2)*(np.cos(6*np.pi*x)+np.sin(6*np.pi*x)))
f2 = lambda h: (-np.exp(-x*2)*(np.cos(6*np.pi*x)+np.sin(6*np.pi*x)))
```

```python
x = np.arange(0,5,0.001)
plt.title("Gráfico")
plt.xlabel("x")
plt.ylabel("y")
plt.grid("on")
plt.plot(x,f(x),linestyle = 'solid',linewidth=3)
plt.plot(x,f2(x),linestyle = 'solid',linewidth=3)
plt.axis(xmin = 0,xmax = 5,ymin =-1,ymax= 1)
plt.show()
```
